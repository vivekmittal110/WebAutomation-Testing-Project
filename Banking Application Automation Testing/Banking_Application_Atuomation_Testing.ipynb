```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

# Set up WebDriver
driver = webdriver.Chrome()  # Ensure ChromeDriver is installed and in PATH
driver.maximize_window()

try:
    # Open the OrangeHRM demo login page
    driver.get("https://opensource-demo.orangehrmlive.com/web/index.php/auth/login")

    # Wait for the username field to load
    time.sleep(5)  # Waiting for the page to load (you can adjust this duration)
    
    # Enter the username
    driver.find_element(By.XPATH, "//input[@name='username']").send_keys("Admin")

    # Wait for the password field to load
    time.sleep(1)  # Short wait before interacting with the password field
    
    # Enter the password
    driver.find_element(By.XPATH, "//input[@type='password']").send_keys("admin123")

    # Wait for the login button to be ready
    time.sleep(1)  # Short wait before clicking the login button
    
    # Click the login button
    driver.find_element(By.XPATH, "//button[@type='submit']").click()

    # Wait for the dashboard header to be visible (increased time or WebDriverWait)
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//h6[contains(., 'Dashboard')]"))
    )

    # If the element is found, login is successful
    print("Login successful")

except NoSuchElementException:
    print("Login failed. Dashboard header not found.")
except Exception as e:
    print("An error occurred:", str(e))

finally:
    # Close the browser after a short wait
    time.sleep(6)  # Wait before closing the browser to view results
    driver.quit()
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options

# Path to your Edge WebDriver (update this path to your local WebDriver location)
edge_driver_path = "C:/Users/91931/Downloads/edgedriver_win64/msedgedriver.exe"

# Set up WebDriver options for Edge
edge_options = Options()
edge_options.add_argument("--start-maximized")  # Maximize the window
edge_options.add_argument("--disable-extensions")  # Disable extensions if needed

# Set up Edge WebDriver
service = Service(edge_driver_path)
driver = webdriver.Edge(service=service, options=edge_options)

try:
    # Open the OrangeHRM demo login page
    driver.get("https://opensource-demo.orangehrmlive.com/web/index.php/auth/login")

    # Wait for the username field to load
    time.sleep(5)  # Waiting for the page to load (you can adjust this duration)
    
    # Enter the username
    driver.find_element(By.XPATH, "//input[@name='username']").send_keys("Admin")

    # Wait for the password field to load
    time.sleep(2)  # Short wait before interacting with the password field
    
    # Enter the password
    driver.find_element(By.XPATH, "//input[@type='password']").send_keys("admin123")

    # Wait for the login button to be ready
    time.sleep(2)  # Short wait before clicking the login button
    
    # Click the login button
    driver.find_element(By.XPATH, "//button[@type='submit']").click()

    # Wait for the dashboard header to be visible (increased time or WebDriverWait)
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//h6[contains(., 'Dashboard')]"))
    )

    # If the element is found, login is successful
    print("Login successful")

except NoSuchElementException:
    print("Login failed. Dashboard header not found.")
except Exception as e:
    print("An error occurred:", str(e))

finally:
    # Close the browser after a short wait
    time.sleep(5)  # Wait before closing the browser to view results
    driver.quit()
```
